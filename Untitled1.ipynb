```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("zahrazolghadr/ab-test-cookie-cats")

print("Path to dataset files:", path)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import math

df = pd.read_csv("/content/cookie_cats.txt")

# Group analysis
summary = df.groupby("version")[["retention_1", "retention_7"]].mean()
print(summary)

# Chi-square test
table_1 = pd.crosstab(df["version"], df["retention_1"])
table_7 = pd.crosstab(df["version"], df["retention_7"])

chi2_1, p1, _, _ = chi2_contingency(table_1)
chi2_7, p7, _, _ = chi2_contingency(table_7)

print("Day 1 p-value:", p1)
print("Day 7 p-value:", p7)

# Confidence interval
def ci(p, n):
    z = 1.96
    se = math.sqrt((p * (1 - p)) / n)
    return (p - z * se, p + z * se)

for v in ["gate_30", "gate_40"]:
    subset = df[df["version"] == v]
    p = subset["retention_7"].mean()
    n = len(subset)
    print(v, ci(p, n))
```

📊 Cookie Cats A/B Test Analysis

Two game versions were compared: gate_30 and gate_40. The analysis showed that Day 1 retention was very similar between the groups: 44.8% for gate_30 and 44.2% for gate_40. The p-value for Day 1 retention was 0.0755, which is greater than 0.05, indicating that the difference is not statistically significant.

For Day 7 retention, the results showed a clearer difference: 19.0% for gate_30 versus 18.2% for gate_40. The p-value was 0.0016, which is below the 0.05 threshold, meaning the difference is statistically significant. This suggests that the gate_30 version performs better in retaining players over a longer period.

The confidence intervals also support this conclusion, as the retention range for gate_30 is consistently slightly higher than for gate_40. Overall, moving the gate from level 30 to level 40 had a negative impact on long-term player retention.
